## Outliner Treatment

```python
uv=np.percentile(df.n_hot_rooms,[99])[0] #99% percnetile value of n_hot_rooms
uv
```

```python
df[(df.n_hot_rooms>uv)]
```

```python
df.n_hot_rooms[(df.n_hot_rooms)>3*uv]=3*uv
```

* rainfall have outliner but on lower side so will will replace it with the [0.3*1-percentile]

```python
lv=np.percentile(df.rainfall, [1])[0]
lv
```

```python
df.rainfall[(df.rainfall<lv)]
```

```python
df[(df.rainfall<0.3*lv)]=0.3*lv
```

## Train test split

```python
X_multi = df.drop("price", axis=1)
X_multi.head()
```

```python
y_multi = df['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_multi, y_multi, test_size= 0.2, random_state=0)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
ml3 = LinearRegression()
```

```python
ml3.fit(X_train, y_train)
```

```python
y_test_a = ml3.predict(X_test)
```

```python
y_train_a = ml3.predict(X_train)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_test_a)
```

```python
r2_score(y_train, y_train_a)
```
