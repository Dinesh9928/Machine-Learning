## Missing Values Imputation

```python
df.Time_taken.fillna(value = df['Time_taken'].mean(), inplace=True)
```

```python
df.describe()
```

## Dummy variable creation

```python
df = pd.get_dummies(df, columns = ['3D_available','Genre'], drop_first=True)
```

```python
df.head()
```

## X-y Split

```python
X = df.loc[:, df.columns !='Collection']
```

```python
X.head()
```

```python
y = df['Collection']
```

```python
y.head()
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

## Standardizing the data 

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler().fit(X_train)
```

```python
X_train_s = sc.transform(X_train)
```

```python
X_test_s = sc.transform(X_test)
```

```python
X_train_s
```

## Training Regression SVM Model

```python
from sklearn.svm import SVR
```

```python
svr_1 = SVR(kernel='linear', C=350)
```

```python
svr_1.fit(X_train_s, y_train)
```

## Predicting Values 

```python
y_train_pre = svr_1.predict(X_train_s)
y_test_pre = svr_1.predict(X_test_s)
```

## Evaluating Performance

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
mean_squared_error(y_train, y_train_pre)
```

```python
r2_score(y_train,y_train_pre)
```

```python
r2_score(y_test, y_test_pre)
```
