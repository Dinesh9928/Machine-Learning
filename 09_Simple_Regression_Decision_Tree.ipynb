# DECISION REGRESSION TREE

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('F:/Data Science/Machine Learning Udemy/Course Material/Data Files/3. ST Academy - Decision Trees resource files/Movie_regression.csv', header=0)
```

## Missing values 

```python
df['Time_taken'].fillna(value = df.Time_taken.mean(), inplace=True)
```

```python
df.info()
```

# Dummy Variable Creation

```python
df= pd.get_dummies(df, columns=['3D_available', 'Genre'], drop_first=True)
```

```python
df.head()
```

# X-y Split

```python
X = df.loc[:, df.columns!="Collection"]
type(X)
```

```python
y = df['Collection']
type(y)
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size=0.2, random_state=0)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

# Training Regression Tree

```python
from sklearn import tree
reg_tree = tree.DecisionTreeRegressor(max_depth=3)
```

```python
reg_tree.fit(X_train, y_train)
```

# Predicting Values Using Trained Data

```python
y_train_pred = reg_tree.predict(X_train)
y_test_pred = reg_tree.predict(X_test)
```

# Evaluating The Performance

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
print(f'MSE for training data {mean_squared_error(y_train, y_train_pred)} and for test data {mean_squared_error(y_test, y_test_pred)}')
```

```python
print(f'r2_Score\ntraining data {r2_score(y_train, y_train_pred)}\ntest data {r2_score(y_test, y_test_pred)}')
```

# Ploting The Decision Tree

```python
dot_data = tree.export_graphviz(reg_tree, out_file=None)
```

```python
from IPython.display import Image
```

## Maximum number of level in Tree

```python
regtree1 = tree.DecisionTreeRegressor(max_depth=3)
regtree1.fit(X_train, y_train)
dot_data = tree.export_graphviz(regtree1,out_file=None,feature_names=X_train.columns, filled = True)
graph1 = pydotplus.graph_from_dot_data(dot_data)
Image(graph1.create_png())
```

## Minimum Observation at Internal Nodes

```python
regtree2 = tree.DecisionTreeRegressor(min_samples_split= 50)
regtree2.fit(X_train, y_train)
dot_data = tree.export_graphviz(regtree2, feature_names=X_train.columns, filled = True)
graph2 = pydotplus.graph_from_dot_data(dot_data)
Image(graph2.create_png())
```

## Minimum Observation at Leaf Nodes

```python
regtree2 = tree.DecisionTreeRegressor(min_samples_leaf=25)
regtree2.fit(X_train, y_train)
dot_data = tree.export_graphviz(regtree2, feature_names=X_train.columns, filled = True)
graph2 = pydotplus.graph_from_dot_data(dot_data)
Image(graph2.create_png())
```

## More than one condition on tree

```python
regtree2 = tree.DecisionTreeRegressor(min_samples_leaf=20, max_depth=4, min_samples_split=25)
regtree2.fit(X_train, y_train)
dot_data = tree.export_graphviz(regtree2, feature_names=X_train.columns, filled = True)
graph2 = pydotplus.graph_from_dot_data(dot_data)
Image(graph2.create_png())
```
