# Ensemble Techiques

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('F:/Data Science/Machine Learning Udemy/Course Material/Data Files/3. ST Academy - Decision Trees resource files/Movie_classification.csv', header = 0)
```

```python
df.head()
```

```python
df.describe()
```

## Missing value Imputation

```python
df.Time_taken.fillna(value = df['Time_taken'].mean(), inplace=True)
df.info()
```

## Dummy Variable Creation

```python
df = pd.get_dummies(df, columns = ['3D_available','Genre'], drop_first=True)
```

```python
df.head()
```

## X-y Split

```python
X = df.loc[:, df.columns!='Start_Tech_Oscar']
```

```python
y = df['Start_Tech_Oscar']
```

```python
X.head()
```

```python
y.head()
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

## Classification Tree Creation

```python
from sklearn import tree
```

```python
clftree = tree.DecisionTreeClassifier( max_depth= 7)
```

```python
clftree.fit(X_train, y_train)
```

## Value Prediction

```python
y_test_pred = clftree.predict(X_test)
y_train_pred = clftree.predict(X_train)
```

## Model Performance

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
confusion_matrix(y_train, y_train_pred)
```

```python
accuracy_score(y_train, y_train_pred)
```

```python
confusion_matrix(y_test, y_test_pred)
```

```python
accuracy_score(y_test, y_test_pred)
```

## Plotting Decision Tree

```python
dot_data = tree.export_graphviz(clftree, out_file=None,feature_names=X_train.columns,filled=True)
```

```python
from IPython.display import Image
import pydotplus
```

```python
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

```python
clftree2 = tree.DecisionTreeClassifier(max_depth=4, min_samples_leaf=25)
clftree2.fit(X_train, y_train)
dot_data = tree.export_graphviz(clftree2, out_file=None,feature_names=X_train.columns,filled=True)
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
accuracy_score(y_test, clftree2.predict(X_test))
```

# Bagging

```python
from sklearn import tree
clftree = tree.DecisionTreeClassifier()
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
bg_clf = BaggingClassifier( n_estimators=500,bootstrap=True, n_jobs=-1, random_state=42)
```

```python
bg_clf.fit(X_train, y_train)
```

```python
confusion_matrix(y_test, bg_clf.predict(X_test))
```

```python
accuracy_score(y_test, bg_clf.predict(X_test))
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
ran_clf = RandomForestClassifier(n_estimators=1000, n_jobs=-1, random_state=42)
```

```python
ran_clf.fit(X_train, y_train)
```

```python
confusion_matrix(y_test, ran_clf.predict(X_test))
```

```python
accuracy_score(y_test, ran_clf.predict(X_test))
```

# Grid Search CV

```python
from sklearn.model_selection import GridSearchCV
```

```python
param ={
    'max_features': [3,4,5,6,7,8,9,10,12,14],
    'min_samples_split': [2,3,5,10]
}
```

```python
rn_clf = RandomForestClassifier(n_estimators=250, n_jobs=-1, random_state=42)
```

```python
grid_search = GridSearchCV(rn_clf, param, n_jobs=-1, cv=5, scoring='accuracy')
```

```python
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
cvrf_clf = grid_search.best_estimator_
```

```python
confusion_matrix(y_test, cvrf_clf.predict(X_test))
```

```python
accuracy_score(y_test, cvrf_clf.predict(X_test))
```

## Grid search CV

```python
xgb_clf1 = xgb.XGBClassifier(n_estimator = 1000, learning_rate = 0.08, random_state = 42 )
```

```python
param_test = {
    'max_depth': range(2,10,2),
    'gamma': [0.1,0.2,0.3],
    'subsample': [0.8,0.9,1],
    'colsample_bytree': [0.8,0.9,1],
    'reg_alpha': [1e-2, 0.1,0.8,1]
}
```

```python
grid_searchxgb = GridSearchCV(xgb_clf1, param_test,n_jobs = -1, cv=5, scoring ='accuracy')
```

```python
grid_searchxgb.fit(X_train, y_train)
```

```python
grid_searchxgb.best_params_
```

```python
xgb_best = grid_searchxgb.best_estimator_
```

```python
accuracy_score(y_test, xgb_best.predict(X_test))
```
